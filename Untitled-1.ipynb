### Modelado, Inversion e Integracion Geofisica

### Modelo directo

En el modelado directo conocemos la salida del modelo matematico que representa el fenomeno de estudio asignando parametros conocidos y un dominio en especifico

Refiriendonos al modelo matematico mas sencillo tenemos: $$y=m_1+m_2x$$
Donde:
- $y$ es la salida del modelo
- $m_1, m_2$ los parametros del modelo
- $x$ es el dominio de observacion

___

### Modelo inverso

Es el **calculo de los parametros** a partir de la salida del **modelo matematico** o **los datos observados en campo**
- Gravimetria: densidad, profundidad y forma de estructuras
- Sismica: modelos de densidad y velocidad y parametros de Lame
- Magnetometria: susceptibilidades y forma de estructuras
- Electrica: resistividades

Menke define los parametros como propiedades intrinsecas de la materia.

La ecuacion de onda: $$\frac{{\partial ^2 u}}{{\partial x^2 }} = \frac{1}{{c^2 }}\frac{{\partial ^2 u}}{{\partial t^2 }}$$

El problema de inversion es lineal cuando los parametros se comportan de manera lineal, cuando solo soportan una derivada.
Cuando nos enfrentamos a problemas de inversion no lineal, lo primero es intentar linealizar el problema: $$y=e^{m_1x}$$
Problema linealizado: $$ln(y)=m_1x$$
Cuando no es posible linealizar el problema se utiliza metodo de inversion no lineal como: algoritmos geneticos, perturbaciones aleatorias, recristalizacion simulada que hacen uso de procesos probabilisticos y fuerza computacional.

Para realizar la inversion lineal se utilizan procesos estadisticos con diferentes criterios en funcion del objetivo del problema.


**Tipos de sistemas lineales**
Un sistema es deterministico cuando la solucion es unica, es es que el proceso de inversion no es estadistico, ejemplo sistema de ecuaciones de 2x2

El sistema lineal lo podemos representar de la forma $$d=Gm$$esto es que podemos realizar una operacion matriz-vector. $G$ contiene el dominio y el modelo matematico que modela el fenomeno, $d$ es la salida, y $m$ los parametros.
La solucion del sistema es: $$m=dG^{-1}$$


En un **sistema indeterminado** tenemos mas incognitas que numero de ecuaciones por lo que el numero de soluciones es infinito

En un sistema sobre determinado es cuando tenemos mas ecuaciones que parametros (incognitas). Y tenemos que asignar un criterio para dar solucion al problema inverso.
El primer criterio para dar solucion al sistema sobredeterminado es proponer una funcion que aproxime la mayor cantidad de puntos observados, buscando $$d^{obs}=d^{cal}$$ para medir la similitud entre 2 vectores hacemos uso del concepto de distancia euclidiana.
En casos ideales en que el modelo matematico describe perfectamente el fenomeno en estudio y los datos observados estan libres de ruidos tenemos que $$d^{obs}-d^{cal}=\bar0$$
En la naturaleza tenemos:$$d^{obs}-d^{cal}=\bar{e}$$
donde $\bar{e}$ es el vector de errores.

Para medir la similitud entre 2 vectores hacemos uso del concepto de distancia euclidiana (norma L2) $$E={\sum\limits_{i=1}^{n} {ei^2}}$$

Existe una familia de normas definidas donde basadas en la potencia elevada a la raiz obtenida. Donde las normas pequenas le dan preferencias a la mayoria de los datos, las normas altas le dan preferencia a los valores extremos.


Podemos expresar la norma L2 de la siguiente manera: 
El vector de errores es un vector columna de tamano n*1 donde n es el numero de observaciones 
$$e=\begin{bmatrix}e_1\\ e_2\\ ...\\ e_n\end{bmatrix}$$ 
$$E=e'(1*n)e(n*1)$$
$$E=[e_1,e_2,...,e_n] \begin{bmatrix}e_1\\ e_2\\ ...\\ e_n\end{bmatrix} = \sum\limits_{i=1}^{n} {{e_i}^2}$$
$$Gm-d=e$$
$$E=e'-e$$
$$E=(Gm-d)'(Gm-d)$$


Optimizamos la funcion E respecto a los parametros
$$E=(m'G'-d')(Gm-d)$$
$$E=m'G'Gm-m'G'd-d'Gm-d'd$$
$$\frac{{\partial E}}{{\partial m'}} = G'Gm-G'd$$
$$G'Gm-G'd=0$$
$$G'Gm=G'd$$
$$m=(G'G)^{-1}G'd $$
Solucion para un sistema indeterminado

Construimos la expresion $Gm=d$ a partir del sistema lineal para una 
$$y_1=m_1+m_1 x_1$$ 
$$y_2=m_1+m_2 x_2$$ 
$$...$$ 
$$y_5=m_5+m_5 x_5$$
$$\begin{bmatrix}y_1\\ y_2\\ ...\\ y_5\end{bmatrix}=
\begin{bmatrix}1 & x_1\\1 & x_2\\... & ...\\ 1 & x_5 \end{bmatrix}
\begin{bmatrix}m_1 & m_2\end{bmatrix}$$
El problema directo consiste en calcular $d$ conociendo $G$ y $m$ . El problema inverso consistes en encontrar $m$ a partir de $d$ y $G$.
Para sistemas deterministicos la solucion es: $$m=G^{-1}d$$
Para el caso de sistemas lineales sobredeterminados realizamos un proceso de optimizacion, donde la condicion para optimizar el sistema esta descrita por la distancia minima entres 2 vectores, esto es la distancia minima entre el vector $d^{obs}$ y  el vector salida del modelo matematico $d^{cal}$.
Funcion objetivo o funcion de costo: contiene restricciones asociadas a los procesos de inversion.<br>
Entonces la funcion de costo se expresa como (funcion de costo): $$\phi = E$$
$$\phi =e'e$$
$$\phi = (Gm-d)'(Gm-d)$$
$$\phi = (m'G'Gm-m'G'd-d'Gm-d'd)$$
$$\frac{{\partial \phi}}{{\partial m'}} = G'Gm-G'd$$
$$G'Gm-G'd=0$$
$$G'Gm=G'd$$
$$m=(G'G)^{-1}G'd $$
Solucion sistema indeterminado <br>
Para este caso en particular $G'(2x5)G(5x2)$ esta multiplicacion siempre va a ser una matriz cuadrada del numero de parametros de inversion. En este caso (2x2)

Solucion por MMC convencionales para un sistema indeterminado 
$$y_1=m_1+m_1 x_1$$ 
$$y_2=m_1+m_2 x_2$$ 
$$...$$
$$y_5=m_5+m_5 x_5$$
$$\begin{bmatrix}y_1\\ y_2\\ ...\\ y_5\end{bmatrix}=
\begin{bmatrix}1 & x_1\\1 & x_2\\... & ...\\ 1 & x_5 \end{bmatrix}
\begin{bmatrix}m_1 & m_2\end{bmatrix}$$

Donde la solucion esta dada por $$m=(G'G)^{-1}G'd$$
Donde:
$$G'G=\begin{bmatrix}1 & 1 & ... & 1\\ x_1 & x_2 & ... & x_n\end{bmatrix}
\begin{bmatrix}1 & x_1\\ 1 & x_2\\ ... & ... \\ 1 & x_n \end{bmatrix}=
\begin{bmatrix}N & \sum\limits_{i=1}^{N} {x_i} \\ \sum\limits_{i=1}^{N} {x_i} & \sum\limits_{i=1}^{N} {{x_i}^2}\end{bmatrix}$$
$$G'd=\begin{bmatrix}1 & 1 & ... & 1\\ x_1 & x_2 & ... & x_n\end{bmatrix}
\begin{bmatrix}y_1\\ y_2\\ ...\\ y_n\end{bmatrix}=
\begin{bmatrix} \sum\limits_{i=1}^{N} {y_i}\\ \sum\limits_{i=1}^{N} {x_i y_i}\end{bmatrix}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

## Straight line inversion

### Se construye la linea recta

```python
dx=0.5 #dominio de observacion
x=np.arange(0,20,dx)
y=5+7*x+np.random.uniform(1,2,(x.size,))*5
data=pd.DataFrame({'x':x, 'y':y})
data.to_csv('data_line.csv', index=False)
plt.plot(x,y,'or')
plt.show()
```

### Inversion forma desarrolada

```python
data=pd.read_csv('data_line.csv')
x2=data.x.values
y2=data.y.values
#Creamos G'G
N=x2.size
sumx=np.sum(x2)
sumx2=np.sum(x2**2)
GTG=np.array([[N, sumx],[sumx, sumx2]])


#Creamos GTd
sumy=np.sum(y2)
sumxy=np.sum(x2*y2)
Gtd=np.array([[sumy],[sumxy]])
m=np.linalg.solve(GTG,Gtd)

#el modelo matematico esta descrito por
yf=m[0]+m[1]*x2
plt.plot(x2,y2,'go')
plt.plot(x2.T,yf,'r')
plt.show()
```

### Inversion forma implicita

```python
data=pd.read_csv('data_line.csv')
x2=data.x.values
y2=data.y.values

#se construye G, y se desarrolla m=inv(G'G)(Gd)
G=np.array((np.ones(x2.size).T, x2.T))
m = np.linalg.inv(G.dot(G.T)).dot(G.dot(y2))

#el modelo matematico esta descrito por
yf=m[0]+m[1]*x2

fig=go.Figure()
fig.add_trace(go.Scatter(x=x2,y=y2, mode='markers'))
fig.add_trace(go.Scatter(x=x2,y=yf, mode='lines'))
fig.show()

#plt.plot(x2,y2,'go')
#plt.plot(x2.T,yf,'r')
#plt.show()
```

## Parabola inversion

### Creamos la parabola

```python
dx=0.1 #dominio de observacion
x=np.arange(0,20,dx)
y=5+7*x+3*x**2;
data=pd.DataFrame({'x':x, 'y':y})
data.to_csv('data_parabola.csv', index=False)
plt.plot(x,y,'.r')
plt.show()
```

### Inversion forma desarrollada

```python
data=pd.read_csv('data_parabola.csv')
x2=data.x.values
y2=data.y.values

#Creamos G'G
N=x2.size
sumx=np.sum(x2)
sumx2=np.sum(x2**2)
sumx3=np.sum(x2**3)
sumx4=np.sum(x2**4)
GTG=np.array([[N, sumx, sumx2],[sumx, sumx2, sumx3],[sumx2, sumx3, sumx4]])

#Creamos GTd
sumy=np.sum(y2)
sumxy=np.sum(x2*y2)
sumx2y=np.sum(x2**2*y2)
Gtd=np.array([[sumy],[sumxy],[sumx2y]])
m=np.linalg.solve(GTG,Gtd)

#el modelo matematico esta descrito por y=m1+m2x+m3x^2
yf2=m[0]+m[1]*x2+m[2]*(x**2)
plt.plot(x2,y2,'ro')
plt.plot(x2,yf2,'b')
plt.show()
```

### Inversion forma implicita

```python
data=pd.read_csv('data_parabola.csv')
x2=data.x.values
y2=data.y.values

#se construye G, y se desarrolla m=inv(G'G)(Gd)
G=np.array((np.ones(x2.size).T, x2.T, x2.T**2))
m = np.linalg.inv(G.dot(G.T)).dot(G.dot(y2))

#el modelo matematico esta descrito por
yf=m[0]+m[1]*x2+m[2]*x2**2
plt.plot(x2,y2,'ro')
plt.plot(x2.T,yf,'b')
plt.show()
```

## Plane surface inversion

### Creamos el plano

```python
# modelo m1+m2x+m3y

[X,Y]=np.meshgrid(np.arange(0,20,0.1),np.arange(10,30,0.2))
P=5+7*X+9*Y
data=pd.DataFrame({'x':X.flatten(), 'y':Y.flatten(), 'p':P.flatten()})
data.to_csv('data_plane.csv', index=False)

fig=go.Figure(data=[go.Surface(x=X, y=Y, z=P)])
fig.show()
```

### Inversion implicita

```python
data=pd.read_csv('data_plane.csv')
x2=data.x.values
y2=data.y.values
p2=data.p.values

X2=np.reshape(x2,(100,200))
Y2=np.reshape(y2,(100,200))
P2=np.reshape(p2,(100,200))

G=np.array((np.ones(p2.size), x2, y2))
m = np.linalg.inv(G.dot(G.T)).dot(G.dot(p2))

PF=m[0]+m[1]*X2+m[2]*Y2
fig=go.Figure(data=[go.Surface(x=X2, y=Y2, z=PF)])
fig.show()
```

## Straigth line with Lagrage

### Se crea la linea recta

```python
dx=0.5 #dominio de observacion
x=np.arange(0,20,dx)
y=5+7*x+np.random.uniform(1,2,(x.size,))*5
data=pd.DataFrame({'x':x, 'y':y})
data.to_csv('data_line_lagrange.csv', index=False)
plt.plot(x,y,'or')
plt.show()
```

### Inversion implicita

```python
data=pd.read_csv('data_line_lagrange.csv')
x2=data.x.values
y2=data.y.values

#definimos la restriccion
xr1, yr1= np.array(15), np.array(90)

#se construye G y F
G=np.array((np.ones(x2.size).T, x2.T))
F=np.array([1,xr1])

#construimos el modelo m=inv(A)B
A1=np.concatenate((G.dot(G.T),np.array([F.T])), axis=0)
A2=np.concatenate((F, np.zeros(xr1.size)),axis=0)
A=np.concatenate((A1.T, np.array([A2])), axis=0)

B1=np.array([G.dot(y2.T)])
B=np.append(B1,yr1) 
''' append por que solo es un valor para este ejemplo, 
supongo que si se aumenta el numero de restricciones puede que siga funcionando'''

m = np.linalg.inv(A).dot(B)

#el modelo matematico esta descrito por
yf=m[0]+m[1]*x2
plt.plot(x2,y2,'go')
plt.plot(xr1,yr1,'bo')
plt.plot(x2.T,yf,'r')
plt.show()
```

## Parabola with Lagrange

### Creamos la parabola

```python
dx=0.1 #dominio de observacion
x=np.arange(0,20,dx)
y=5+7*x+3*x**2;
data=pd.DataFrame({'x':x, 'y':y})
data.to_csv('data_parabola_lagrange.csv', index=False)
plt.plot(x,y,'.r')
plt.show()
```

### Inversion implicita

```python
data=pd.read_csv('data_parabola_lagrange.csv')
x2=data.x.values
y2=data.y.values

#se definen las resticciones
xr=np.array([1,12])
yr=np.array([50,375])

#se construye G
G=np.array((np.ones(x2.size).T, x2.T, x2.T**2))
F=np.concatenate((np.array([np.ones(xr.size)]),np.array([xr.T]),([np.array(xr**2).T]))).T

A1=np.concatenate((G.dot(G.T),F), axis=0)
A2=np.concatenate((F.T, np.zeros((xr.size,xr.size))),axis=0)
A=np.hstack((A1, A2)) # tambien puede ser usado np.concatenate((A1, A2), axis=1)

B1=np.array([G.dot(y2.T)])
B=np.append(B1,yr) 
''' append por que solo es un valor para este ejemplo, 
supongo que si se aumenta el numero de restricciones puede que siga funcionando'''

m = np.linalg.inv(A).dot(B)


#el modelo matematico esta descrito por
yf=m[0]+m[1]*x2+m[2]*x2**2
plt.plot(x2,y2,'ro')
plt.plot(xr, yr,'bo')
plt.plot(x2.T,yf,'b')
plt.show()
```
